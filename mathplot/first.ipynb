# Matplotlib Learning

```python
import matplotlib.pyplot as plt
```

```python
# line chat

x = [1,2,3,4,5]
y = [1,2,3,4,1]

plt.figure(figsize=(4,3))

plt.plot(x,y)
plt.show()
```

## Design the graph

```python
plt.plot(x,y, color = 'pink' , marker = 'o')
plt.xlabel("X ")
plt.ylabel("Y ")
plt.show()
```

## Multiply graphh

```python
y1 = [4,5,6,7,8]
y2= [41,52,63,73,84]

plt.plot(x,y1, label="Sales 22", color="pink" )
plt.plot(x,y2, label="Sales 23", color="blue")

plt.Figure(figsize=(3,2))

plt.title("Sales in XYZ Company")
plt.xlabel("Months")
plt.ylabel("Sales")
plt.legend()
plt.show()

```

## Bar Charts

```python
plt.bar( ['A' ,'B' , 'CC' ,'DD' , "FF"] , y1, color = "pink")
plt.title("Bar charts")
plt.xlabel("Months")
plt.ylabel("Days")
plt.show()
```

## Histograph

```python
xx = [3,4,3,4,3,4,34,34,4,34,43,4,3,4,3,43,4,3,4,23]
plt.hist(xx,bins=2)
plt.title("Histograph")
plt.show()
```

## Pie Chart

```python

plt.pie(x, labels=y , autopct='%1i%%')
plt.show()
```
